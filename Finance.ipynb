```python
# Define tickers for Magnificent 7
tickers_mag7 = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'NVDA', 'META', 'TSLA']
```

```python
# Download adjusted close prices
data = yf.download(tickers_mag7, start='1995-01-01')['Close']
```

```python
# Calculate simple daily returns
daily_returns = data.pct_change()
```

```python
plt.figure(figsize=(12, 6))
for t in tickers_mag7:
    daily_returns[t].plot(label=t, alpha=0.6)
plt.title("Daily Returns of Magnificent 7 Stocks")
plt.xlabel("Date")
plt.ylabel("Return")
plt.legend()
plt.show()
```

```python
# Calculate annualized average return
avg_returns = daily_returns.mean() * 250
# Display formatted annualized returns
for t in tickers_mag7:
    print(f"{t}: {round(avg_returns[t]*100, 3)}%")
```

```python
# Define tickers
mag7_etf = 'MAGS'   # Replace with correct ticker if different
spy_etf = 'SPY'
```

```python
# Set date range (last 10 years)
end_date = datetime.today()
start_date = end_date - timedelta(days=10*365)
```

```python
data = yf.download([mag7_etf, spy_etf], start=start_date, end=end_date)['Close']
data.dropna(inplace=True)
```

```python
# Normalize prices to start at 1
normalized = data / data.iloc[0]
```

```python
# Plot normalized returns
plt.figure(figsize=(10, 6))
normalized[mag7_etf].plot(label=f'{mag7_etf} (MAG7 ETF)')
normalized[spy_etf].plot(label='SPY (S&P 500)')
plt.title('10-Year Total Return Comparison: MAG7 ETF vs SPY')
plt.xlabel('Date')
plt.ylabel('Normalized Price (Start = 1)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Cumulative return
cumulative_return = (data.iloc[-1] / data.iloc[0]) - 1
print("Cumulative Returns (last 10 years):")
print(cumulative_return.apply(lambda x: f"{x:.2%}"))
```

```python
# Annualized return
annualized_return = (data.iloc[-1] / data.iloc[0]) ** (1/10) - 1
print("\nAnnualized Returns:")
print(annualized_return.apply(lambda x: f"{x:.2%}"))
```

```python
# === Top 50 tickers provided ===
tickers_top50 = [
    'NVDA', 'MSFT', 'AAPL', 'AMZN', 'META', 'AVGO', 'GOOGL', 'GOOG', 'BRK-B', 'TSLA',
    'JPM', 'WMT', 'LLY', 'V', 'ORCL', 'NFLX', 'MA', 'XOM', 'COST', 'JNJ', 'PG', 'HD',
    'BAC', 'ABBV', 'PLTR', 'KO', 'PM', 'CSCO', 'UNH', 'GE', 'WFC', 'IBM', 'CVX',
    'TMUS', 'CRM', 'AMD', 'ABT', 'MS', 'AXP', 'LIN', 'DIS', 'GS', 'MCD', 'MRK',
    'INTU', 'UBER', 'NOW', 'TXN', 'T', 'RTX'
]
# === Add SPY for benchmark ===
all_tickers = tickers_top50 + ['SPY']
# === Date range ===
end_date = datetime.today()
start_date = end_date - timedelta(days=10 * 365)
```

```python
data = yf.download(all_tickers, start=start_date, end=end_date, auto_adjust=True)['Close']
data = data.dropna(axis=0)  # Drop dates with missing prices for any ticker
```

```python
# Daily returns
returns = data.pct_change().dropna()

# Equal-weighted return of top 50
top50_return = returns[tickers_top50].mean(axis=1)

# SPY return
spy_return = returns['SPY']

cumulative_returns = (1 + returns[tickers_top50]).cumprod()
years = 10
final_values = cumulative_returns.iloc[-1]
annualized_returns = final_values ** (1 / years) - 1

# Convert to DataFrame and sort
performance_df = pd.DataFrame({
    'Cumulative Return': final_values - 1,
    'Annualized Return': annualized_returns
})

performance_df_sorted = performance_df.sort_values('Annualized Return', ascending=False)
print(performance_df_sorted)

performance_df_sorted['Annualized Return'].head(10).plot(kind='bar', figsize=(10,5), title='Top 10 Stocks by Annualized Return')
plt.ylabel('Annualized Return')
plt.tight_layout()
plt.show()
```
